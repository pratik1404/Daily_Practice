```python
x = [1, 2, 3, 4, 5]
y1 = [10,20,30,40,50]
y2 = [21,32,43,54,64]

plt.figure(figsize = (9,9))

plt.subplot(2, 2, 1)
plt.plot(x , y1 , color = "red" , linestyle = "dotted" , linewidth = 2 , marker = "v")
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("first plot")

plt.subplot(2, 2, 2)
plt.plot( y1,x  ,color = "green" , linestyle = "dotted" , linewidth = 2 , marker = "^")
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("second plot")

plt.subplot(2, 2, 3)
plt.plot( x, y2 ,color = "yellow" , linestyle = "dotted" , linewidth = 2 , marker = "s")
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("third plot")

plt.subplot(2, 2, 4)
plt.plot( y2,x  ,color = "blue" , linestyle = "dotted" , linewidth = 2 , marker = "p")
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("fourth plot")
```

### Histogram

* Histogarams are used to represent the distribution of a dataset . They divide the data into buns and count the number of data points in each bin. 

| Feature         | **Histogram** 🟦📊                         | **Bar Chart** 🟩📉                          |
| --------------- | ------------------------------------------ | ------------------------------------------- |
| **Use**         | To show **frequency distribution** of data | To compare **different categories**         |
| **X-Axis**      | Shows **ranges of numbers (bins)**         | Shows **categories/labels**                 |
| **Bars Touch?** | ✅ **Yes**, bars are **touching**           | ❌ **No**, bars are **separated**            |
| **Data Type**   | **Continuous data** (e.g., marks, height)  | **Categorical data** (e.g., fruits, cities) |
| **Example Use** | Student marks distribution                 | Sales per product                           |

```python

# Data
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_scores = [40, 45, 55, 60, 70, 80, 85]

# Create Scatter Plot
plt.scatter(study_hours, exam_scores, color='blue', marker='o')

# Add Labels and Title
plt.title('Study Hours vs. Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score (%)')

# Show Grid
plt.grid(True)

# Display the plot
plt.show()
```

```python

import numpy as np

# Generate large dataset (1000+ points)
np.random.seed(42)
experience = np.random.uniform(0, 20, 1500)  # 1500 data points from 0 to 20 years
salary = 2 * experience + np.random.normal(0, 3, 1500)  # Linear trend + noise

# Plot the large scatter plot
plt.figure(figsize=(9, 7))
plt.scatter(experience, salary, color='purple', alpha=0.5, s=20)

# Add title and labels
plt.title('Experience vs. Salary (Large Dataset)', fontsize=16)
plt.xlabel('Experience (Years)', fontsize=12)
plt.ylabel('Salary (LPA)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()

```

```python
## pie chart

label = ["a" , "b" , "c" , "d" , "e"]
sizes = [30,40,50,60 ,70]
color = ["gold" ,"magenta","teal" , "lightcoral" , "lightskyblue"]
explode= (0.2,0,0,0,0)

plt.pie(sizes, labels=label , colors = color )
```

```python
## pie chart and visualize one slide

label = ["a" , "b" , "c" , "d" , "e"]
sizes = [30,40,50,60 ,70]
color = ["gold" ,"magenta","teal" , "lightcoral" , "lightskyblue"]
explode= (0.2,0,0,0.2,0)

plt.pie(sizes ,labels=label ,explode=explode , colors = color , autopct = "%1.1f%%" , shadow = True)
```
